## Skeletonized flourish

```python
%matplotlib notebook
skimage.io.imshow(
    skimage.morphology.binary_dilation(
        skimage.morphology.skeletonize(skimage.color.rgb2gray(flourish)>.6), 
        skimage.morphology.disk(5)
    ));
```
