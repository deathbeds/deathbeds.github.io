Getting text directly from an image.

```python
    import jason, pytesseract, IPython
```

```python
    class Img(metaclass=jason.Stype):
        def load(str): return pytesseract.image_to_string(__import__('PIL').Image.open(str))
```

```python
    print(Img%'tess_sample.png')
```
