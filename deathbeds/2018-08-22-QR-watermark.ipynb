# A QR Code "Watermark"
You could use this while giving a presentation, or in a kiosk display.

```python
%reload_ext importnb
from jinja2 import Template
from IPython.utils.capture import capture_output
import qrcode
from base64 import b64encode
from IPython.display import HTML
from deathbeds.__qr_codes_are_magic import qr
```

```python
def qr_watermark(uri="https://deathbeds.github.io/", tag=None, opts="-o svg"):
    if tag is None:
        tag = f"""<iframe src="{uri}" width="100%" height="100%" style="border: 0; background-color: #fff; min-height: 60vh"></iframe>"""
    with capture_output() as cap:
        %qr $uri $opts
    raw_bytes = cap.outputs[0].data['image/svg+xml'].encode("utf-8")
    img_uri = f"data:image/svg+xml;base64,{b64encode(raw_bytes).decode('utf-8')}"
    return f"""
        <div style="position: relative;">
            {tag}
            <img src="{img_uri}" style="position: absolute; right: 1em; bottom: 1em;"/>
        </div>
    """
```

```python
if __name__ == "__main__":
    display(HTML(qr_watermark()))
```
