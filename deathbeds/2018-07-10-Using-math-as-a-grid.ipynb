We'll use __sympy__ and some simple functions to design a basis for writing the word __DEATHBEDS__.  The functions will be plotted and projected onto a wall to trace out the guidelines for the top, bottom, and middle of the letters.

```python
    from sympy.plotting import *
    from sympy import *
    from ipywidgets import interact
    import matplotlib.pyplot as plt
    plt.rcParams['figure.figsize'] = 10, 5

    from IPython.display import display, Image
```

## Designing the guidelines

The border function encapsulates our simple guidelines.

![](data/tracing.jpg)

```python
    ω, x, Δ = symbols('ω, x, Δ')
    top, bottom, middle = cos(ω*x) + Δ*0, cos(ω*x) + Δ, cos(ω / 2 * x) + sin(ω / 2 * x) + Δ/2
```

* Create an interactive widget to decide on the guidelines.

```python
    def guidelines(Ω=(1, 20, 0.1), δ=(1, 25, 0.1)):
        plot(*map(
            lambda x: x.subs(ω, Ω).subs(Δ, δ), (top, bottom, middle)
        ), (x, -pi, pi))
    interact(guidelines);
```

![](data/two_worlds.jpg)

* Below are the final settings for Ω & δ.

```python
    Ω, δ = 3.8, 5.9

    guidelines(Ω, δ)
```
