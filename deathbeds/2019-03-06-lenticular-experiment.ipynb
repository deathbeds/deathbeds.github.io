```python
    dx = 3
    h = 800
    w = 900
    frames = 6
    dwg = svgwrite.Drawing(
        height=str(h), width=str(w), debug=True, 
        onmouseover="""document.getElementById('ggg').setAttribute(
        'style', 'clip-path: url(#mask)'
        )""",
        onmouseout="""document.getElementById('ggg').setAttribute(
        'style', ''
        )""",
        onmousemove="""document.getElementById('mask').setAttribute(
        'transform', 'translate('+(this.getBoundingClientRect().left-event.offsetX)+', 0)'
        )""")
    
    mask = dwg.defs.add(dwg.clipPath(id='mask'))
    mask.translate(2)
    for i in range(-2*w, 2*w, dx*frames):
        mask.add(dwg.rect((i, 0), (dx, h)))

    group = dwg.add(dwg.g(id="ggg"))
    colors = "pink turquoise blue blueviolet violet lavender".split()
    for frame in range(frames):
        color = colors[frame]
        clip_path = dwg.defs.add(dwg.clipPath(id=color))
        group.add(dwg.rect((frame*dx, 0), (w-frame*dx, h),  style=F"clip-path: url(#{color});").fill(color))
        for i in range(0, w, dx*frames):
            clip_path.add(dwg.rect((frame*dx + i, 0), (dx, h)))
            
    dwg
```
